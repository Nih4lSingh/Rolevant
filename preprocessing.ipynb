```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
custom_stopwords = stop_words.union({
    'experience', 'work', 'years', 'skills', 'ability', 
    'strong', 'team', 'role', 'position', 'job', 'company',
    'opportunity', 'required', 'looking', 'must', 'will',
    'also', 'well', 'new', 'good', 'great', 'high', 'including',
    'responsible', 'working', 'related', 'knowledge', 'excellent',
    'time', 'ensure', 'support', 'within', 'across', 
    'provide', 'benefits', 'us', 'one', 'etc', 'based',
    'join', 'help', 'build', 'develop', 'manage', 'need',
    'use', 'make', 'get', 'also', 'well', 'plus',
    'sleep', 'smile', 'beauty', 'gaming', 'resort',
    'unlimited', 'career', 'opportunities', 'pay',
    'bonus', '000', '1st', '7am', 'snf',
    'ulta', 'aspen', 'ingersoll', 'rand', 'edward', 
    'jones', 'oak', 'street', 'hca', 'apex', 'navy',
    'capital', 'white', 'cap', 'allied', 'universal',
    'eisner', 'epsilon', 'ctdi', 'davita', 'petco',
    'dnv', 'skf', 'skanska',
})
```

```python
import re
def clean_text(combined):
    combined=combined.lower()
    combined=re.sub(r'[^\x00-\x7F]+',' ',combined)
    combined = re.sub(r'[^\w\s]', ' ', combined)
    words=combined.split()
    filtered=[word for word in words if word not in custom_stopwords]
    return ' '.join(filtered)
```

```python
role_texts_clean = df_clean.groupby('role')['combined'].apply(lambda x: ' '.join(x.dropna()))
role_texts_clean = role_texts_clean.apply(clean_text)
```

```python
from sentence_transformers import SentenceTransformer
model=SentenceTransformer('all-mpnet-base-v2')
```

```python
def get_embedding(text):
    return model.encode(text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=15000,ngram_range=(1,3),min_df=3)
tfidf_matrix = vectorizer.fit_transform(role_texts_clean.values)
feature_names = vectorizer.get_feature_names_out()
```

```python
import numpy as np

def get_top_keywords_for_role(role_index, top_n=30):
    row = tfidf_matrix[role_index].toarray()[0]
    top_indices = row.argsort()[::-1][:top_n]
    return [feature_names[i] for i in top_indices]
```

```python
role_keywords = {}
for i, role in enumerate(role_texts_clean.index):
    row = tfidf_matrix[i].toarray()[0]
    top_indices = row.argsort()[::-1][:30]
    role_keywords[role] = feature_names[top_indices]
```

```python
c=1
for role, keywords in list(role_keywords.items())[:121]:
    print(c)
    c+=1
    print(f"\n{role}:")
    print(keywords)
```

```python
role_embeddings={}
for role,text in role_texts_clean.items():
    role_embeddings[role]=get_embedding(text)
```

```python
roles={}
for role in role_keywords.keys():
    roles[role]={
        'keywords': role_keywords[role],
        'embeddings': role_embeddings[role].tolist()
    }
```

```python
import json
import numpy as np

class NumpyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        if isinstance(obj, np.floating):
            return float(obj)
        if isinstance(obj, np.integer):
            return int(obj)
        return super().default(obj)

with open('role_data.json', 'w') as f:
    json.dump(roles, f, cls=NumpyEncoder)

print("Saved successfully")
```
