```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
```

```python
from transformers import pipeline
```

```python
skill_extractor= pipeline(
    'token-classification',
    model='algiraldohe/lm-ner-linkedin-skills-recognition',
    aggregation_strategy='simple'
)
```

```python
role_texts_clean = df_clean.groupby('role')['combined'].apply(lambda x: ' '.join(x.dropna()))
```

```python
from sentence_transformers import SentenceTransformer
model=SentenceTransformer('all-mpnet-base-v2')
```

```python
def get_embedding(text):
    return model.encode(text)
```

```python
from groq import Groq
from kaggle_secrets import UserSecretsClient
import ast
from collections import Counter
import re

secret = UserSecretsClient()
GROQ_API_KEY = secret.get_secret("GROQ_API_KEY")

client = Groq(api_key=GROQ_API_KEY)
```

```python
def extract_skills_from_role(role_name,text,chunk_size=5):
    sentences= sent_tokenize(text)
    chunks=[]
    for i in range(0,len(sentences),chunk_size):
        chunk=' '.join(sentences[i:i+chunk_size])
        chunks.append(chunk)
        
    all_skills=[]
    for chunk in chunks:
        results=skill_extractor(chunk)
        skills= [result['word'] for result in results
        if result['score']>=0.9]
        all_skills.extend(skills)
    skill_counts = Counter(all_skills)
    filtered_skills = [skill for skill, count in skill_counts.items() if count >= 2]
    prompt=f"""
You are a career expert. Given this list of extracted keywords for the role of {role_name}, 
remove anything that is NOT a genuine required skill for this role.
Keep only actual technical skills, tools, methodologies, and relevant soft skills.
Remove industry context words, company names, benefits, and irrelevant terms.

Keywords: {filtered_skills}

Return ONLY a valid Python list of strings.
Do not include explanations, notes, markdown, or any additional text.
Example:
["Python", "SQL", "Machine Learning"]
Output MUST be of the above format
"""
    
    response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "system",
         "content": "You are a skill extraction assistant. You only output valid Python lists. Never add explanations, comments, or any text outside the list brackets."}
        ,{
            "role":"user",
            "content":prompt
          }
        ]
    )
    raw=response.choices[0].message.content.strip()
    match= re.search(r'\[.*?\]',raw,re.DOTALL)   
    return ast.literal_eval(match.group())
```

```python
import time
import numpy as np

role_skills={}
count=47
for role, text in role_texts_clean.items():
    print(f"Processing: {role}")
    role_skills[role]=extract_skills_from_role(role,text)
    count-=1
    print(f"Done {count} is left")
    time.sleep(3)
```

```python
role_skills={role:list(set(skills)) for role,skills in role_skills.items()}
```

```python
def compute_embedding(text):
    sentences=sent_tokenize(text)
    embedding= [model.encode(sentence) for sentence in sentences]
    return np.mean(embedding,axis=0)
```

```python
role_embeddings={}
count=47
for role, text in role_texts_clean.items():
    print(f"Encoding: {role}")
    role_embeddings[role]=compute_embedding(text)
    count-=1
    print(f"Done {count} is left")
    
```

```python
role_data={}
for role in role_skills.keys():
    role_data[role]={
        "keywords":role_skills[role],
        'embeddings': role_embeddings[role].tolist()
    }
```

```python
import json

class NumpyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        if isinstance(obj, np.floating):
            return float(obj)
        if isinstance(obj, np.integer):
            return int(obj)
        return super().default(obj)

with open('role_data_v3.json','w') as f:
    json.dump(role_data,f,cls=NumpyEncoder)
print("Saved sucessfully")
```
